```python
import matplotlib.pyplot as plt
import numpy as np
```

# Matlab Way of Life

```python
x = np.linspace(0, 10, 100) # criamos a figure
plt.figure() # criamos o primeiro subplot 

# realizamos o plot no primeiro subplot
plt.subplot(2, 1, 1) # (rows, columns, painel number)
plt.plot(x, np.sin(x))


# criamos o segundo subplot
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

# Object Oriented

```python
fig, ax = plt.subplots(2) # indicamos que queremos 2 subplots

ax[0].plot(x, np.sin(x));
ax[1].plot(x, np.cos(x));
```

```python
people = ['Ann','Brandon','Chen','David','Emily','Farook',
'Gagan','Hamish','Imran','Julio','Katherine','Lily']
age = [21,12,32,45,37,18,28,52,5,40,48,15]
weight = [55,35,77,68,70,60,72,69,18,65,82,48]
height = [160,135,170,165,173,168,175,159,105,171,155,158]
```

```python
%config InlineBackend.figure_format='retina'
```

```python
fig, ax = plt.subplots(figsize=(12,9))
ax.scatter(age, height)
ax.legend(['Altura em cm'], loc='upper left', fontsize=14)

ax.set_title('Idade vs Altura (cm)', fontsize=22, fontstyle='italic')
ax.set_xlabel('Idade (anos)', fontsize=16)
ax.set_ylabel('Altura (cm)', fontsize=16)
ax.set_xticks([i*5 for i in range(13)])
ax.tick_params(axis="x", labelsize=12)
ax.tick_params(axis="y", labelsize=12)
ax.set_ylim(100,200)
ax.grid(True)
ax.scatter(age, height, c='orange', s=150, edgecolor='black')

ax.text(x=22, y=185, s="Nobody's higher than 180cm")
ax.text(x=15, y=120, rotation=30, s="Just another \ncomment")
ax.hlines(y=180, xmin=0, xmax=40, linestyle='dashed', color='r', lw=3)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.set_title("Peso dos individuos (Kg)")
ax.bar(x=people, height=weight, width=0.6, color='orange', edgecolor='k', alpha=0.6)
ax.set_xlabel('Indivíduos')
ax.set_ylabel('Peso (Kg)')
ax.tick_params(axis='x', labelsize=12, rotation=30)
fig.set_facecolor('seashell')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1, y2, y3 = np.cos(x), np.cos(x+1), np.cos(x+3)
names = ['Signal 1', 'Signal 2', 'Signal 3']

fig, ax = plt.subplots(nrows=3,figsize=(10,8))
fig.tight_layout(pad=3.0)
ax[0].set_facecolor('seashell')
ax[1].set_facecolor('seashell')
ax[2].set_facecolor('seashell')

ax[0].plot(x, y1);
ax[0].set_title("Signal 1");

ax[1].plot(x, y2);
ax[1].set_title("Signal 2");

ax[2].plot(x, y3);
ax[2].set_title("Signal 3");
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10,8))

names = ['Signal 1', 'Signal 2', 'Signal 3']

for i , j, k in zip(axes, (y1, y2, y3), names):
    print(i, j, k)
```
