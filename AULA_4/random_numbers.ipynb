# RANDOM WALK

```python
import random
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20,8) # definindo o tamanho padrao de todas as imagens

position = 11
my_walk = [position]

for i in range(1000):
    if random.uniform(0,1) > 0.44:
        step = 1
    else:
        step = -1
    position = position + step
    my_walk.append(position)
    
plt.plot(my_walk[:1000]);
```
