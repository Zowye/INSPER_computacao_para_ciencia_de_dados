# Creating a personalized Histogram from random int library

```python
import numpy as np
import pandas as pd
my_values = np.random.randint(low=0,
                              high=24,
                              size=2000)

df_teste = pd.DataFrame(my_values, columns=['horario'])

df_teste.head(4)
```

```python
values = np.random.normal(loc=12, scale=6.0, size=1000)

df_teste = pd.DataFrame(values, columns=['horario'])

max_value = int(df_teste.max())
min_value = int(df_teste.min())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'axes.facecolor':'white'}) #para dark mode no vscode
fig, ax1 = plt.subplots(nrows=1,figsize=(10,8))
sns.histplot(data=df_teste, x="horario", binwidth=1, bins=max_value, ax = ax1, kde=True)

# Forcing changing x-ticks values
ax1.tick_params(axis="x", labelrotation=5, labelsize=12)

# Changing ticks formatting
ax1.xaxis.set_ticks( np.arange(min_value, max_value, 1) )
ax1.set_xlabel(xlabel='Horário Local', labelpad=15)
ax1.xaxis.label.set_fontsize(20)
ax1.set_ylabel(ylabel='Contagem', labelpad=15)
ax1.yaxis.label.set_fontsize(20)
```
