# COVID

```python
url = 'https://corgis-edu.github.io/corgis/datasets/json/covid/covid.json'
covid = requests.get(url).json()
```

```python
covid = requests.get(url).json()
```

```python
df_covid = pd.read_json(url)
```

```python
covid_date = pd.json_normalize(df_covid['Date'])
covid_data = pd.json_normalize(df_covid['Data'])
covid_location = pd.json_normalize(df_covid['Location'])
```

```python
df_concated = pd.concat([covid_date, covid_data, covid_location], axis=1)
```

```python
df_concated['Date'] = pd.to_datetime(df_concated[['Year', 'Month', 'Day']])
```

```python
top_10 = df_concated.groupby('Country')['Cases'].sum().reset_index().sort_values(by='Cases', ascending=False).head(10)
```

```python
top_10_countries = list(top_10['Country'])
```

```python
df_top_10_countries = df_concated[df_concated['Country'].isin(top_10_countries)]
```

```python
df_top_10_countries.head(2)
```

```python
a = alt.Chart(df_top_10_countries).mark_line().encode(
    x = alt.X('Date', title='Date'),
    y = alt.Y('Cases', title='Cases'),
    color = alt.Color('Country'),
    tooltip=['Date', 'Cases', 'Country']
).properties(width=800, height=500)
```

```python
df_top_10_countries['CumSum'] = df_top_10_countries.groupby(['Country'])['Cases'].cumsum()
```

```python
b = alt.Chart(df_top_10_countries).mark_line(size=2).encode(
    x = alt.X('Date', title='Date'),
    y = alt.Y('CumSum', title='CumSum'),
    color = alt.Color('Country'),
    tooltip=['Date', 'CumSum', 'Country']
).properties(width=800, height=500)
```
